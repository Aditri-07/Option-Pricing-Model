**THE BLACK SCHOLES MERTON MODEL**

black scholes formula.png

```python
from math import sqrt
import numpy as np
from scipy.stats import norm
```

```python
import pandas as pd
data = pd.read_csv("option_GS_df.csv", parse_dates=["Expiration", ' DataDate'])
data
```

```python
#difference in days between Expiration and DataDate
data['date_diff'] = (data['Expiration'] - data[' DataDate']).dt.days

#the spread between Ask and Bid prices
data['Spread'] = data['Ask'] - data['Bid']

data.head(10)
```

```python
#filter
data_call = data[data['Type'] == 'call'].reset_index(drop=True)
print(data_call.head(10))
```

**Black-Scholes Model Implementation**

1. S is the stock price
2. K is the strike price
3. T is the fraction of days to strike date divided by 252
4. r is the risk free rate
5. sigma is the annual volatility of the returns of the stock

```python
#define the Black-Scholes model for European call options
def black_scholes(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    #put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return call_price
```

**Applying Black Scholes Model to the Dataset**

```python
#function to calculate modeled prices for all call options
def modelled_prices(data, r=0.0056):
    data['m_prices'] = 0
    for i in range(len(data)):
        S = data.loc[i, 'UnderlyingPrice']
        K = data.loc[i, 'Strike']
        T = data.loc[i, 'date_diff'] / 252
        sigma = data.loc[i, 'IVMean']
        data.at[i, 'm_prices'] = black_scholes(S, K, T, r, sigma)
    return data
```

```python
final = modelled_prices(data_call) #apply the function
print(final[['Bid', 'm_prices']].head(10))
```

```python
import matplotlib.pyplot as plt #plot

plt.figure(figsize=(10, 5))
plt.scatter(final['Bid'], final['m_prices'], c='royalblue', label='Bid vs. Model Prices')
plt.plot([final['Bid'].min(), final['Bid'].max()], [final['Bid'].min(), final['Bid'].max()], 'r--',label='y=x')
m, b = np.polyfit(final['Bid'], final['m_prices'], 1)
plt.plot(final['Bid'], m*final['Bid'] + b, color='green', label='Line of Best Fit')
plt.xlabel('Bid Prices')
plt.ylabel('Modeled Prices')
plt.title('Bid Prices vs. Modeled Prices for Call Options')
plt.legend()
plt.show()
```
